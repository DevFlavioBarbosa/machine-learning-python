```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/DevFlavioBarbosa/machine-learning-python/main/conjuntos_de_dados/02_01_tiny_data.csv')
```

```python
df.head()
```

```python
colunas_valores = [col for col in df.columns if col != 'produtos']
```

```python
colunas_valores
```

```python
df_long = pd.melt(
    df,                        # DataFrame original
    id_vars=['produto'],      # Coluna que permanecerá como identificador
    value_vars=colunas_valores,  # Colunas que serão transformadas
    var_name='valor',          # Nome da nova coluna que armazenará os nomes das colunas originais
    value_name='qnt_produto'   # Nome da nova coluna que armazenará os valores
)
```

```python
df_long.sort_values(['produto', 'valor'])
```
