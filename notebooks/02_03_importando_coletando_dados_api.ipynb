# Importando e coletando dados usando API

Aprenda a consumir dados através do uso de API.

```python
import requests
import pandas as pd
```

```python
indicador = requests.get('https://servicodados.ibge.gov.br/api/v1/paises/BR/indicadores/77830').json()
```

```python
indicador
```

```python
df_indicador = pd.DataFrame.from_dict(indicador[0]['series'][0]['serie'])
```

```python
df_limpo = df_indicador.dropna(axis=1, how='all')
```

```python
for col in df_limpo.columns:
    df_limpo[col] = pd.to_numeric(df_limpo[col], errors='coerce')
```

```python
mask_tem_valores = df_limpo.notna().any(axis=0)
```

```python
df_final = df_limpo.loc[:, mask_tem_valores]
```

```python
df_final.sample(10)
```

```python
# Coletar todos os valores não nulos de todas as colunas
valores_reais = []
nomes_colunas = []

for coluna in df_final.columns:
    valores_nao_nulos = df_final[coluna].dropna()
    for valor in valores_nao_nulos:
        valores_reais.append(valor)
        nomes_colunas.append(f"{coluna}_{len([x for x in valores_reais if str(x).startswith(str(coluna))])}")

# Criar dataframe com uma única linha
df_uma_linha = pd.DataFrame([valores_reais], columns=nomes_colunas)
```
